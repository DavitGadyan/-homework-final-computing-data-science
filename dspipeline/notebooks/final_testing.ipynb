```python
from dspipeline.data.load import Data
from dspipeline.spliting.split import Split
from dspipeline.processing.prep import DropNaNs, FillNaNs
from dspipeline.processing.transformer import Dummifier, Binarizer, Logger
from dspipeline.modeling.model import Model
from dspipeline.evaluating.evaluate import Metric
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *
```

```python
X_FEATURES = ['Tenure', 'CityTier', 'WarehouseToHome', 'HourSpendOnApp',
       'NumberOfDeviceRegistered', 'SatisfactionScore', 'NumberOfAddress',
       'Complain', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount',
       'DaySinceLastOrder', 'CashbackAmount', 'Churn',
       'PreferredLoginDevice_Computer', 'PreferredLoginDevice_Mobile Phone',
       'PreferredLoginDevice_Phone', 'PreferredPaymentMode_CC',
       'PreferredPaymentMode_COD', 'PreferredPaymentMode_Cash on Delivery',
       'PreferredPaymentMode_Credit Card', 'PreferredPaymentMode_Debit Card',
       'PreferredPaymentMode_E wallet', 'PreferredPaymentMode_UPI',
       'PreferedOrderCat_Fashion', 'PreferedOrderCat_Grocery',
       'PreferedOrderCat_Laptop & Accessory', 'PreferedOrderCat_Mobile',
       'PreferedOrderCat_Mobile Phone', 'PreferedOrderCat_Others',
       'MaritalStatus_Divorced', 'MaritalStatus_Married',
       'MaritalStatus_Single', 'M/F']
Y_FEATURE = 'Churn'
```

#### Load the data. c. Train test split

```python
data = Data(filename='../data/E Commerce Dataset.xlsx', tab_alias='E Comm')
input_data = data.from_excel()
splitter =  Split()
X_y_train, X_y_test =  splitter.split_train_test(data=input_data, target=Y_FEATURE)

print('X_y_train>>', X_y_train)
print('X_y_test>>', X_y_test)
```

## a. Preprocessing: Remove those rows that contain NaN values in the columns

```python
X_y_train = DropNaNs(X_y_train, ['HourSpendOnApp', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder']).drop()
X_y_test = DropNaNs(X_y_test, ['HourSpendOnApp', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder']).drop()

print('X_y_train>>', X_y_train.info())
print('X_y_test>>', X_y_test.info())
```

## a. Fill NaN with the mean value of the column in the columns

```python
print('X_y_train nans before filling>>>\n', X_y_train.isna().sum().T[['Tenure', 'WarehouseToHome',]])
print('X_y_test nans before filling>>>\n', X_y_test.isna().sum().T[['Tenure', 'WarehouseToHome', ]])

X_y_train = FillNaNs(X_y_train, ['Tenure', 'WarehouseToHome', ]).fill()
X_y_test = FillNaNs(X_y_test, ['Tenure', 'WarehouseToHome', ]).fill()

print('X_y_train nans after filling>>>\n', X_y_train.isna().sum().T[['Tenure', 'WarehouseToHome', ]])
print('X_y_test nans after filling>>>\n', X_y_test.isna().sum().T[['Tenure', 'WarehouseToHome', ]])
```

## b. Feature generation

```python
X_y_train.info()
```

```python
# dummify ethnicity
X_y_train = Dummifier(X_y_train, ['PreferredLoginDevice', 'PreferredPaymentMode', 'PreferedOrderCat', 'MaritalStatus']).generate()
X_y_test = Dummifier(X_y_test, ['PreferredLoginDevice', 'PreferredPaymentMode', 'PreferedOrderCat', 'MaritalStatus']).generate()

print('X_y_train>>', X_y_train.info())
print('X_y_test>>', X_y_test.info())
```

```python
X_y_train.describe()
```

```python
# binarize gender
X_y_train = Binarizer(X_y_train, 'Gender', 'M/F').generate()
X_y_test = Binarizer(X_y_test, 'Gender', 'M/F').generate()

print('X_y_train>>', X_y_train.info())
print('X_y_test>>', X_y_test.info())
```

```python
# log a variable
X_y_train = Logger(X_y_train, 'Tenure').generate()
X_y_test = Logger(X_y_test, 'Tenure').generate()

print('X_y_train>>', X_y_train.info())
print('X_y_test>>', X_y_test.info())
```

## d. Train model & generate predictions

```python
X_y_train.columns
```

```python
model_ = Model(feature_columns=X_FEATURES, target_column=Y_FEATURE, model=LogisticRegression, hyperparamaterers={'penalty': 'l2', 'solver': 'saga'})
# model_ = Model(feature_columns=X_FEATURES, target_column=Y_FEATURE, model=RandomForestClassifier, hyperparamaterers={'n_estimators': 1000, 'criterion': 'gini'})
```

```python
model_.train(X_y_train)
preds = model_.predict_proba(X_y_test)
```

## e. Evaluation

```python
metric = Metric()

metric.calculate_score(score_list=['accuracy', 'precision', 'recall', 'roc_auc'], y_true=X_y_test["Churn"], y_preds=preds)
```
